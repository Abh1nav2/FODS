```python
#8
def check(x):
  if x>0:
    print("If case,x is positive")
  elif x<0:
    print("Elif case,x is negetive")
  else:
    print("Else case,x is zero")
check(5)
check(-5)
check(0)
x=0
print("-----------While loop-------------")
while(x<10):
  x+=1
  if x==5:
    print("Continue when x =5")
    continue
  if x==9:
    print("break when x=9")
    break
  print(x)
print("-----------For loop---------------")

for i in range(10,20):
  if i==15:
    print("Continue when i=15")
    continue
  if i==18:
    print("break when i=18")
    break
  print(i)
```
