# Hypothesis testing

## H0: Branch C makes an averge total of 337 Dollars per day.
## H1: Branch C makes more than 337 Dollars per day.

```python
sample=df.loc[df['Branch']=='C']
samplec=sample.sample(165)
samplec
```

```python
samplec.Total.mean()
```

## z test

```python
#standard deviation of the population
sd=np.std(samplec['Total'])
sd
```

```python
samplec.std()
```

```python
xbar=samplec.Total.mean()
mu=337.09
n=165 #sample size
alpha=0.05
```

```python
xbar
```

```python
zvalue=(xbar-mu)/(sd/np.sqrt(n))
zvalue
```

```python
z_critical = st.norm.ppf(alpha)  #One tail Test
z_critical
```

```python
if (zvalue<z_critical and abs(zvalue)> abs(z_critical)):
  print("Reject Null Hypotheisis")
else:
  print("Fail to reject Null Hypothesis")
```

## using p value

```python
p_val = (1 - st.norm.cdf(abs(zvalue)))
p_val
```

```python
#comparing with alpha value
if p_val<alpha:
  print("Reject Null Hypotheisis")
else:
  print("Fail to reject Null Hypothesis")
```

## Hence Branch C On Average doesn't make more than 337$ per day, i.e, we fail to reject the Null Hypothesis

## T test

```python
tsamplec=sample.sample(25)
tsamplec
```

```python
t_sd=np.std(tsamplec['Total'])
t_sd
```

```python
xbar=tsamplec.Total.mean()
print(xbar)
mu=337.09
n=25 #sample size
alpha=0.05
```

```python
n = 25
degrees_of_freedom = n-1
t = (xbar - mu)/(t_sd / np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha, degrees_of_freedom)
t_critical
```

```python
if (t > t_critical):
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
```

## Both Z test and T test suggest the same result, The Null Hypothesis cannot be rejected.
