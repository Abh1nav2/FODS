```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/reisanar/datasets/master/BostonHousing.csv')
df
```

```python
df=df.rename(columns={'CAT_MEDV':'CAT__MV'})
df
```

```python
df.describe()
```

Scatter Plot

```python
df.plot.scatter(x="LSTAT",y="MEDV")
```

Bar Graphs

```python
plt.bar(df['AGE'],df['LSTAT'])
```

```python
#bar graph plotted for classes
#have to make classes first

cat1=df.groupby("CHAS").mean().MEDV.plot(kind="bar")
cat1.set_ylabel("Avg_MEDV")
cat1.set_xlabel("CHAS_Catergories")
```

Histogram

```python
hist=df.MEDV.hist()
hist.set_xlabel("MEDV")
hist.set_ylabel("Count")
```

Box Plot

```python
box=df.boxplot(column="MEDV",by="CHAS")
plt.suptitle("")
plt.title("")
```

```python
x,y=plt.subplots(nrows=1,ncols=4)
df.boxplot(column="NOX",by="CAT__MV",ax=y[0])
df.boxplot(column="LSTAT",by="CAT__MV",ax=y[1])
df.boxplot(column="PTRATIO",by="CAT__MV",ax=y[2])
df.boxplot(column="INDUS",by="CAT__MV",ax=y[3])
plt.suptitle("")  
plt.tight_layout() 
```
